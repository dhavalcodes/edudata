```python
list_of_columns = ['District Code', 'School Code','County Name', 'District Name', 'School Name', 'Zip Code', 'Test Type', 'Grade', 'Students Enrolled', 'Students Tested', 'Percentage Standard Exceeded', 'Percentage Standard Met',
       'Percentage Standard Met and Above', 'Percentage Standard Nearly Met',
       'Percentage Standard Not Met', 'Students with Scores', 'Test ID']

tested[list_of_columns]

primary_dataset = dataset[list_of_columns]

#Test ID 1 means English
#Test ID 2 means Math
```

```python
primary_dataset['Test ID'].value_counts()
```

```python
english_dataset = primary_dataset[primary_dataset['Test ID'] == 1]
math_dataset = primary_dataset[primary_dataset['Test ID'] == 2]
```

```python
#converting String values to floats
primary_dataset.dtypes

columns_to_convert = [
    "Students Enrolled",
    "Students Tested",
    "Percentage Standard Exceeded",
    "Percentage Standard Met",
    "Percentage Standard Met and Above",
    "Percentage Standard Nearly Met",
    "Percentage Standard Not Met",
    "Students with Scores"
]

primary_dataset[columns_to_convert] = primary_dataset[columns_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
primary_dataset.dtypes
```

```python
english_dataset

def subject_filter(df, subject):
  #df is the dataframe to filter
  #subject is either "math" or "english"
  if subject == "english":
    return df[df['Test ID'] == 1]
  elif subject == "math":
    return df[df['Test ID'] ==2]
  else:
    raise ValueError("Subject must be either 'math' or 'english'")


def grade_filter(df, grade):
  #df is the dataframe to filter
  #grade is the grade to filter for
  grade_filtered = df[df['Grade'] == grade]
  return grade_filtered

high_school_math = grade_filter(subject_filter(primary_dataset, "math"), 11)
high_school_english = grade_filter(subject_filter(primary_dataset, "english"), 11)

#schools that need help with a particular course: standard met + standard exceeded is in the bottom tenth percentile

```

```python
#worst math performing high schools in California
worst_math_performing = high_school_math[high_school_math['Percentage Standard Met and Above'].notnull()].sort_values("Percentage Standard Met and Above", ascending=True)
worst_math_performing.sort_values("Percentage Standard Not Met", ascending=False)
worst_math_performing[worst_math_performing['School Name']=="Glendora High"]
```
